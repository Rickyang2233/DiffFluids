```python
import torch
import matplotlib.pyplot as plt
import numpy as np

from ddpm import DDPM
```

```python
model_small = DDPM(in_channels=1, n_feats=256, betas=[1e-4, 0.02], n_T=500, device='cuda:1', pretrained='../../checkpoint/ddpm_smoke_small.pkl')
model_small.to('cuda:1');
```

```python
cs = torch.tensor([[3.0, 6.0, 6.0], [5.0, 6.0, 6.0], [7.0, 6.0, 6.0]])
x_gen_small = model_small.sample(cs, domain_size=(64, 48))
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
for i in range(3):
    ax[i].imshow(x_gen_small[i, 0].cpu().detach().numpy(), cmap='gray', origin='lower')
    # ax[i].set_title(f'Condition: {cs[i]}')
    ax[i].axis('off')
```
