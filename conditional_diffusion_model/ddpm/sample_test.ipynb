```python
import torch
import matplotlib.pyplot as plt
import numpy as np

from ddpm import DDPM
```

```python
model = DDPM(in_channels=1, n_feats=256, betas=[1e-4, 0.02], n_T=400, device='cuda:1', pretrained='../checkpoint/ddpmunet.pkl')
model.to('cuda:1');
```

```python
cs = torch.tensor([[30.0, 10.0, 5.0], [100.0, 40.0, 30.0], [200.0, 70.0, 15.0]])
x_gen = model.sample(cs)
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
for i in range(3):
    ax[i].imshow(x_gen[i].cpu().numpy().squeeze(), cmap='gray', origin='lower')
    ax[i].set_title(f't = {cs[i, 0]}, (x, y) = ({cs[i, 1]}, {cs[i, 2]})')
    ax[i].axis('off')
```
