```python
import torch
import matplotlib.pyplot as plt

from ddpm import DDPM
```

```python
model = DDPM(in_channels=1, n_feats=256, betas=[1e-4, 0.02], n_T=500, device='cuda:1', pretrained='../../checkpoint/ddpm_mnist_baseline.pkl')
model.to('cuda:1');
```

```python
c = torch.ones(10, dtype=torch.int64) * 4
x_gen = model.sample(10, c)
```

```python
x_gen = x_gen.cpu().detach().numpy()
x_gen.shape
```

```python
fig, ax = plt.subplots(2, 5, figsize=(20, 8))
for i in range(2):
    for j in range(5):
        ax[i, j].imshow(x_gen[i*5+j, 0], cmap='gray')
        ax[i, j].axis('off')
```
