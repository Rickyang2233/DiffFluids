```python
import torch
import matplotlib.pyplot as plt

from ddpm import DDPM
```

```python
model_small = DDPM(in_channels=1, betas=[1e-4, 0.02], n_T=400, device='cuda:1', pretrained='/media/bamf-big/gefan/DiffFluids/checkpointsmoke_small_diffuser.pkl')
model_small.to('cuda:1');
```

```python
x_gen_small = model_small.sample(n_samples=10, size=(64, 48))
```

```python
fig, ax = plt.subplots(2, 5, figsize=(20, 10))
for i in range(2):
    for j in range(5):
        ax[i, j].imshow(x_gen_small[i*5+j, 0].detach().cpu().numpy(), cmap='gray', origin='lower')
        ax[i, j].axis('off')
```
